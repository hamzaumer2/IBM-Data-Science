```python
import yfinance as yf

ticker_symbol = "TSLA"

tesla_data = yf.download(ticker_symbol, period="1d", start="2023-01-01", end="2023-10-10")

tesla_data.reset_index(inplace=True)

tesla_data.to_csv("tesla_stock_data.csv", index=False)

print(tesla_data.head())
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

ticker_symbol = "TSLA"

tesla_data = yf.download(ticker_symbol, period="1d", start="2023-01-01", end="2023-10-10")

tesla_data.reset_index(inplace=True)

def make_graph(data, title):
    plt.figure(figsize=(12, 6))
    plt.plot(data['Date'], data['Close'], label='Tesla Closing Price', color='blue')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Closing Price (USD)')
    plt.legend()
    plt.grid(True)
    plt.show()

graph_title = "Tesla Stock Price (2023)"

make_graph(tesla_data, graph_title)
```
